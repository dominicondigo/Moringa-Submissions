```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
food_data = pd.read_csv("food_poverty_est.csv")
food_data.head()
```

```python
food_data.describe()
```

```python
food_data.columns
```

```python
sns.regplot(x=food_data["Headcount Rate (%)"], y=food_data["Distribution of the Poor (%)"], fit_reg=True)
```

```python
food_data['Headcount Rate (%)'].plot.hist(title='histogram of headcount rate', figsize=(12,6))
```

```python
# bivariate plotting
# scatter plot
food_data[food_data['Headcount Rate (%)'] < 35].sample(30).plot.scatter(x='Headcount Rate (%)', y='Distribution of the Poor (%)',figsize=(12, 6),
                     title='scatter plot of Headcount Rate and distribution of the poor')
```

```python
food_data[food_data['Headcount Rate (%)'] < 35].plot.hexbin(x='Headcount Rate (%)', y='Distribution of the Poor (%)', gridsize=15,figsize=(12,6))
```

```python
food_data.plot.bar(stacked=True,title='stacked plot of food poverty estimate',figsize=(12,6))
```

```python
food_data.plot.area(figsize=(12,6))
```

```python
food_data.plot.line(figsize=(12,6))
```

```python

temp = food_data['Severity of Poverty (%)']

plt.figure(figsize=(12,8))
sns.distplot(temp[~((temp-temp.mean()).abs()>3*temp.std())]);
plt.ylabel("density estimate", fontsize=16)
plt.xlabel('Severity of Poverty', fontsize=16)
plt.title("KDE of Severity of poverty", fontsize=16)
plt.show()
```
